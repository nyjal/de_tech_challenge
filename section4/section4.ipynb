```python
#Import request to call API
import requests
country = 'singapore'
response = requests.get(f"https://api.covid19api.com/total/dayone/country/{country}/status/confirmed")

#Check API is being called successfully
print(response)
response_size = len(response.json())
print(response_size)
```

```python
#Export JSON into list

full_list = []
for x in range(response_size):
    daily_list = []
    daily_list.append(response.json()[x]['Date'][:10])
    daily_list.append(response.json()[x]['Cases'])
    full_list.append(daily_list)
```

```python
import plotly.express as px
import pandas as pd
#Convert list to DF
df = pd.DataFrame (full_list, columns = ['date', 'count'])
#Plot df with plotly
fig = px.line(df, x="date", y="count", title="COVID19 cases in Singapore over time")


#Annotation below are optional, to give more context to the timeline
fig.add_annotation(x='2020-04-07',
            text="Circuit breaker",
            showarrow=True,
            ax = 0,
            ay = -50,
            arrowhead=2)

fig.add_annotation(x='2020-06-01', y=35000,
            text="Phase 1",
            showarrow=True,
            ax = 0,
            ay = -50,
            arrowhead=2)

fig.add_annotation(x='2020-06-19', y=41500,
            text="2020-06-19 | Phase 2",
            showarrow=True,
            ax = 0,
            ay = -100,
            arrowhead=2)

fig.add_annotation(x='2020-12-28', y=58000,
            text="2020-12-28 | Phase 3 ",
            showarrow=True,
            ax = 0,
            ay = -80,
            arrowhead=2)

fig.add_annotation(x='2021-05-08', y=61500,
            text="2021-05-08 | Phase 2",
            showarrow=True,
            ax = 0,
            ay = -100,
            arrowhead=2)

fig.add_annotation(x='2021-05-16', y=61500,
            text="2021-05-16 | Phase 2 Heightened Alert",
            showarrow=True,
            ax = 0,
            ay = -130,
            arrowhead=2)

fig.add_annotation(x='2021-06-14', y=61500,
            text="2021-06-14 | Phase 3 Heightened Alert",
            showarrow=True,
            ax = 0,
            ay = -160,
            arrowhead=2)

fig.add_annotation(x='2021-07-22', y=61500,
            text="2021-05-16 | Phase 2 Heightened Alert",
            showarrow=True,
            ax = 0,
            ay = -180,
            arrowhead=2)

fig.add_annotation(x='2021-08-10', y=65000,
            text="2021-08-10 | Preparatory Stage of Transition",
            showarrow=True,
            ax = 0,
            ay = -75,
            arrowhead=2)

fig.add_annotation(x='2021-09-27', y=93000,
            text="2021-09-27 | Stabilisation Phase",
            showarrow=True,
            arrowhead=2)

fig.add_annotation(x='2021-11-22', y=253000,
            text="2021-11-22 | Transition Phase",
            showarrow=True,
            arrowhead=2)

fig.add_annotation(x='2022-04-26', y=1180000,
            text="2022-04-26 | COVID-19 Resilient Nation Phase",
            showarrow=True,
            arrowhead=2)

fig.show()
```
